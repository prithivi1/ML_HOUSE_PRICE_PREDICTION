```python
def treat_outlier(train,i):
    
    q1 = train.quantile(0.25)
    q3 = train.quantile(0.75)
    
    IQR = q3 - q1
    
    lower_range = q1 - 1.7 * IQR
    upper_range = q3 + 1.7 * IQR
    
    print(" {:20s} {:5s}".format(i,str((train<lower_range).sum()+(train>upper_range).sum())))
#     print(lower_range)
#     print(upper_range)
    
    train_low_index = train[train<lower_range].index
    train_high_index = train[train>upper_range].index
    
#     display(train_high_index)
        
    train[train_low_index] = lower_range
    train[train_high_index] = upper_range
    
    return train
```

```python
for i in lst:
    df_numerical[i] = treat_outlier(df_numerical[i],i)
```

##### ANNOVA TEST

```python
annova_data = pd.DataFrame(columns=['feature1','p-value'])

categorical_features_to_drop = []

for i in df_categorical.columns:
    
    temp = df_categorical[0:1460].copy()
    
    lst = list(temp[i].value_counts().index)
    
    dataframe = pd.concat([temp,y],axis=1)
    
    a = dataframe.groupby(by=i)
    
    var1 = [list(dataframe.iloc[a.indices[val]]['SalePrice']) for val in lst]
    
    f_score,p_value = stats.f_oneway(*var1)
    
    if(p_value<0.05):
        annova_data.loc[len(annova_data.index)] = [i,p_value]
    else:
        categorical_features_to_drop.append(i)
```

##### CHI - SQUARE TEST

```python
def cramers_vcorrected_stat(confusion_matrix):

    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    number_of_rows = confusion_matrix.sum()
    phi2 = chi2 / number_of_rows
    row, key = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((key - 1) * (row - 1)) / (number_of_rows - 1))
    r_corr = row - ((row - 1) ** 2) / (number_of_rows - 1)
    k_corr = key - ((key - 1) ** 2) / (number_of_rows - 1)
    return math.sqrt(phi2_corr / min((r_corr - 1), (k_corr - 1)))
```

##### CROSS VALIDATION

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score,cross_val_predict,cross_validate
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
def model(model,x,y,score='neg_mean_absolute_error'):
    error = cross_val_score(model,x,y,cv=k_fold,n_jobs=1,scoring=score)
    return error
```

##### ENSEMBLE

```python
random_forest.fit(X_train,y_train)
g_boost.fit(X_train,y_train)
xg_boost.fit(X_train,y_train)
```

```python
def ensemble(X_test):
    rgPredict = random_forest.predict(X_test)
    gbpredict = g_boost.predict(X_test)
    xgPredict = xg_boost.predict(X_test)
    lightGBMpredict = model_lgbm.predict(X_test)
    catBoostPredict = catBoost_model.predict(X_test)
    
    totalPredict = rgPredict+gbpredict+xgPredict+lightGBMpredict+catBoostPredict
    
    ans = (rgPredict*(rgPredict/totalPredict)) + (gbpredict*(gbpredict/totalPredict)) + (xgPredict*(xgPredict/totalPredict)) + (lightGBMpredict*(lightGBMpredict/totalPredict)) + (catBoostPredict*(catBoostPredict/totalPredict))
    
    return ans
```

```python
ans = ensemble(X_test)
```

```python
display(metrics.mean_absolute_error(y_test,ans))
```

```python
test_predict = ensemble(test)
```
